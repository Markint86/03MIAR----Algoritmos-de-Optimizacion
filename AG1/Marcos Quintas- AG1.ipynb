### Torres de Hanoi con Divide y vencerás - Mejora con etiqueta de fichas

```python
def find(d, actual):
    for k, v in d.items():
        if v == actual:
            return k

def update_dict(d, actual, final):
    for k, v in d.items():
        if v == actual:
            d[k] = final   
            break
```

```python
def Torres_Hanoi(N, desde, hasta, dict_fichas):
  if N ==1 :
    ficha = find(dict_fichas, desde)
    print(f"Lleva la ficha {ficha} desde" ,desde , " hasta ", hasta )
    update_dict(dict_fichas, desde, hasta)

  else:
    Torres_Hanoi(N-1, desde, 6-desde-hasta, dict_fichas)
      
    ficha = find(dict_fichas, desde)
    print(f"Lleva la ficha {ficha} desde" ,desde , " hasta ", hasta )
    update_dict(dict_fichas, desde, hasta)

    Torres_Hanoi(N-1, 6-desde-hasta  , hasta, dict_fichas)

dict_fichas = {'Azul':1, 'Amarilla':1, 'Roja': 1}
Torres_Hanoi(3, 1 , 3, dict_fichas) #diccionario indicando las fichas (de la menor a la de mayor tamaño) y la posicion de la torre que ocupan
```

```python
#Sucesión_de_Fibonacci
#https://es.wikipedia.org/wiki/Sucesi%C3%B3n_de_Fibonacci
#Calculo del termino n-simo de la suscesión de Fibonacci
def Fibonacci(N:int):
  if N < 2:
    return 1
  else:
    return Fibonacci(N-1)+Fibonacci(N-2)

Fibonacci(5)
```

```python
#Mejora Fibonaaci por metodo iteractivo

def Fibonacci_iteractive(n):
    a,b = 1,1
    for i in range(2,n+1):
        result = a+b
        a = b
        b = result
    return result

Fibonacci_iteractive(5)
```

## Devolución de cambio por técnica voraz

```python
def cambio_monedas(N, SM):
  SOLUCION = [0]*len(SM)   #SOLUCION = [0,0,0,0,..]
  ValorAcumulado = 0

  for i,valor in enumerate(SM):
    monedas =  (N-ValorAcumulado)//valor
    SOLUCION[i] = monedas
    ValorAcumulado = ValorAcumulado + monedas*valor

    if ValorAcumulado == N:
      return SOLUCION


cambio_monedas(15,[25,10,5,1])
```

## N-Reinas por técnica de vueta atrás


```python
def escribe(S):
  n = len(S)
  for x in range(n):
    print("")
    for i in range(n):
      if S[i] == x+1:
        print(" X " , end="")
      else:
        print(" - ", end="")


def es_prometedora(SOLUCION,etapa):
  #print(SOLUCION)
  #Si la solución tiene dos valores iguales no es valida => Dos reinas en la misma fila
  for i in range(etapa+1):
    #print("El valor " + str(SOLUCION[i]) + " está " +  str(SOLUCION.count(SOLUCION[i])) + " veces")
    if SOLUCION.count(SOLUCION[i]) > 1:
      return False

    #Verifica las diagonales
    for j in range(i+1, etapa +1 ):
      #print("Comprobando diagonal de " + str(i) + " y " + str(j))
      if abs(i-j) == abs(SOLUCION[i]-SOLUCION[j]) : return False
  return True



def reinas(N, solucion=[], etapa=0):
  if len(solucion) == 0:
      solucion=[0 for i in range(N)]

  for i in range(1, N+1):
    solucion[etapa] = i
    print()
    print('""""""""""""')
    print(solucion)
    print()
    escribe(solucion)
    print()
    print('""""""""""""')

    if es_prometedora(solucion, etapa):
      if etapa == N-1:
        print("SOLUCION")
        print(solucion)
        print()
        escribe(solucion)
        print()
      else:
        reinas(N, solucion, etapa+1)
    else:
      None

    solucion[etapa] = 0

reinas(4)
```

## Viaje por el rio. Programación dinámica

```python
TARIFAS = [
[0,5,4,3,999,999,999],
[999,0,999,2,3,999,11],
[999,999, 0,1,999,4,10],
[999,999,999, 0,5,6,9],
[999,999, 999,999,0,999,4],
[999,999, 999,999,999,0,3],
[999,999,999,999,999,999,0]
]



################################################################
def Precios(TARIFAS):
################################################################
  #Total de Nodos
  N = len(TARIFAS[0])

  #Inicialización de la tabla de precios
  PRECIOS = [ [9999]*N for i in [9999]*N]
  RUTA = [ [""]*N for i in [""]*N]

  for i in range(0,N-1):
    RUTA[i][i] = i             #Para ir de i a i se "pasa por i"
    PRECIOS[i][i] = 0          #Para ir de i a i se se paga 0
    for j in range(i+1, N):
      MIN = TARIFAS[i][j]
      RUTA[i][j] = i

      for k in range(i, j):
        if PRECIOS[i][k] + TARIFAS[k][j] < MIN:
            MIN = min(MIN, PRECIOS[i][k] + TARIFAS[k][j] )
            RUTA[i][j] = k          #Anota que para ir de i a j hay que pasar por k
        PRECIOS[i][j] = MIN

  return PRECIOS,RUTA
################################################################


PRECIOS,RUTA = Precios(TARIFAS)
#print(PRECIOS[0][6])

print("PRECIOS")
for i in range(len(TARIFAS)):
  print(PRECIOS[i])

print("\nRUTA")
for i in range(len(TARIFAS)):
  print(RUTA[i])

#Determinar la ruta con Recursividad
def calcular_ruta(RUTA, desde, hasta):
  if desde == hasta:
    #print("Ir a :" + str(desde))
    return ""
  else:
    return str(calcular_ruta( RUTA, desde, RUTA[desde][hasta])) +  \
                ',' + \
                str(RUTA[desde][hasta] \
              )

print("\nLa ruta es:")
calcular_ruta(RUTA, 0,6)
```

## Puntos mas cercanos

```python
#1D 
import random

lista1D = [2,6,8,1,34,37,48]
```

```python
#fuerza bruta
def puntos_mas_cercanos_1d(lista):
    n = len(lista)
    min_dist = float('inf')
    par_min = (None, None)

    for i in range(n):
        for j in range(i + 1, n):
            dist = abs(lista[i] - lista[j])
            if dist < min_dist:
                min_dist = dist
                par_min = (lista[i], lista[j])

    return par_min, min_dist
```

```python
puntos_mas_cercanos_1d(lista1D)
```

Los puntos mas cercanos aqui eran 2 y 1, que estan a una distancia de 1. Dado que tenemos 2 bucles y la longitud de la lista es n, el orden computacional es O(n^2)

Para mejorar el caso de fuerza bruta, se propone un algoritmo donde primero se ordene la lista (supondremos que aqui Python usa merge sort o quick sort y que entonces la complejidad es O(n*log(n))) y luego ir comparando la distancia entre pares de vecinos

```python
def puntos_mas_cercanos_1d(lista):
    if len(lista) < 2:
        return None

    nums = sorted(lista)  # O(n log n)

    min_dist = float('inf')
    par_min = (None, None)

    for i in range(len(nums) - 1):
        dist = nums[i + 1] - nums[i]  
        if dist < min_dist:
            min_dist = dist
            par_min = (nums[i], nums[i + 1])

    return par_min, min_dist
```

```python
lista1D = [random.randrange(1,1000) for x in range(1000)]
```

```python
puntos_mas_cercanos_1d(lista1D)
```

La complejidad computacional total aqui es O(n*log(n)) dado que la complejidad de ordenacion de la lista domina sobre la complejidad de la iteraccion (O(n)) para listas grandes. Vemos que claramente es una mejora sobre el algoritmo de fuerza bruta

Ahora vamos a resolver el caso 2D usando divide y venceras, que podriamos haber usado para 1D tambien, aunque no hubiesemos mejorado la complejidad computacional. Este es el problema de los closest pair of points: 

https://cp-algorithms.com/geometry/nearest_points.html

```python
#2D
import math

#funcion para calcular la distancia entre 2 puntos (x1, y1) y (x2, y2) usando la distancia euclidea
def distance(p1, p2):
    return math.sqrt((p1[0] - p2[0]) ** 2 + (p1[1] - p2[1]) ** 2)
```

```python
# funcion para encontrar la distancia minima en torno al centro
def stripClosest(strip, d):
    min_dist = d

    # Ordenar los puntos por su coordenada y
    strip.sort(key=lambda point: point[1])

    # Compare each point in the strip
    for i in range(len(strip)):
        for j in range(i + 1, len(strip)):
            if (strip[j][1] - strip[i][1]) < min_dist:
                min_dist = min(min_dist, distance(strip[i], strip[j]))
            else:
                break

    return min_dist

# Divide y venceras
def minDistUtil(points, left, right):
    
    # Caso de fuerza bruta al que se reduce el algoritmo para 2 o menos puntos
    if right - left <= 2:
        min_dist = float('inf')
        for i in range(left, right):
            for j in range(i + 1, right):
                min_dist = min(min_dist, distance(points[i], points[j]))
        return min_dist

    # Encontrar el punto medio inicial (la mitad del plano)
    mid = (left + right) // 2
    mid_x = points[mid][0]

    #Calcular recursivamnete la distancia minima en la parte izquierda y derecha
    dl = minDistUtil(points, left, mid)
    dr = minDistUtil(points, mid, right)

    d = min(dl, dr) #distancia minima existente entre las 2 mitades

    # Construir
    strip = []
    for i in range(left, right):
        if abs(points[i][0] - mid_x) < d:
            strip.append(points[i])

    # Find the minimum distance in the strip
    stripDist = stripClosest(strip, d)

    return min(d, stripDist)

# funcion para encontrar la distancia mas corta entre 2 puntos en el plano
def minDistance(points):
    n = len(points)

    # Ordenamos los puntos por coordenada x
    points.sort(key=lambda point: point[0])

    return minDistUtil(points, 0, n)
```

```python
lista2D = [(random.randrange(1,1000),random.randrange(1,1000)) for x in range(1000)]
```

```python
minDistance(lista2D)
```

Este algoritmo se basa en divide y venceras. Primero

-Divide el plano a la mitad y va diviendo cada parte recursivamente (divide el array de puntos en 2, y asi sucesivamente despues con cada uno de forma recursiva) tambien a la mitad hasta que nos quedamos con subarrays de solo 2 elementos o menos. Calculamos la distancia minima en cada mitad inicial y sacamos el minimo entre las 2, cuya distancia llamaremos ahora d

-Como se puede dar el caso de que los puntos mas cercanos se encuentren uno en cada uno de las mitades iniciales, hacemos un banda de distancia d (d/2 a cada lado) partiendo del centro donde incluidmos a todos los puntos que haya en esa banda. Luego si hay puntos calculamos la distancia minima entre ellos y la comparamos con d. Escogemos la minima entre las 2

Tiene principalmente complejidad por las divisiones recursivas y comparaciones, lo cual da una complejidad O(n log(n))
